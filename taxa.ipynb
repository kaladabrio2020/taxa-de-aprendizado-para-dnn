#### Power Sheduling

$\large \frac{n_0}{(1 + (\frac{t}{s}))^c}$

```python
def power_schedule(eta_0, s=20, c=1):
    def power(epoch):
        return eta_0/(1+( epoch/s ))**c
    return power
```

```python
iter = np.arange(10)
```

```python
power_1 = power_schedule(0.1,)(iter)
power_2 = power_schedule(0.1,c=2)(iter)
power_3 = power_schedule(0.1,c=3, s=10)(iter)
power_4 = power_schedule(0.1,c=3, s=50)(iter)
```

```python
plt.plot(iter, power_1, '-*', label='c=1')
plt.plot(iter, power_2, '-*', label='c=2')
plt.plot(iter, power_3, '-*', label='c=3 | s=10')
plt.plot(iter, power_4, '-*', label='c=3 | s=50')



plt.title('eta_0=0.1')
plt.legend()
plt.show()
```

Quanto maior o c= maior da taxa de decaimento só que o quanto maior o s irá reduzer essa taxa

#### Exponencial Scheduling

```python
def exponencial_scheduling(eta_0=0.1, s=10):
    def exponencial_fn(epoch):
        return eta_0 * (0.1**(epoch/s))
    return exponencial_fn
```

```python
expo_1 = exponencial_scheduling()(iter)
expo_2 = exponencial_scheduling(s=5)(iter)
expo_3 = exponencial_scheduling(s=15)(iter)
expo_4 = exponencial_scheduling(s=20)(iter)
```

```python
plt.plot(iter, expo_1, '-*', label='s=10')
plt.plot(iter, expo_2, '-*', label='s=5')
plt.plot(iter, expo_3, '-*', label='s=15')
plt.plot(iter, expo_4, '-*', label='s=20')



plt.title('eta_0=0.1')
plt.legend()
plt.show()
```

#### piecise constant

```python
def piewcase_scheduling(eta_0=0.001, s=10):
    def piewcase_fn(epoch):
        if   epoch <= 5:
            return eta_0
        elif epoch > 5 and epoch <= 10:
            return 0.0005
        elif epoch > 10:
            return 0.0001
    return piewcase_fn
```
